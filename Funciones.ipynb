```python
import pandas as pd
```

```python
df_games = pd.read_parquet('data/df_games.parquet')
df_items = pd.read_parquet('data/df_items.parquet')
```

```python
genres = df_games['genres']
```

```python
listaDeGeneros = []
for linea in genres:
  for i in linea:
    if i not in listaDeGeneros:
      listaDeGeneros.append(i)

generos = {'genero':listaDeGeneros}
df_f1 = pd.DataFrame(generos)
```

```python

itemsDeUsuario = itemsDeUsuario[itemsDeUsuario["playtime_forever"] != 0.0]
```

```python
itemsDeUsuario
```

```python
itemsDeUsuario['item_id'] = itemsDeUsuario['item_id'].astype(int)
itemsDeUsuario = itemsDeUsuario.merge(df_games[['id','genres','anio']], left_on='item_id', right_on='id')
itemsDeUsuario
```

```python
df_f1[['anio','Horas jugadas']] = None
```

```python
for i in df_f1['genero'].values:
    genero = i

    #Juegos del genero pedido
    mascara = itemsDeUsuario['genres'].apply(lambda x: genero in x)

    #Calcular las horas por aÃ±o con juegos del genero
    sumaDehoras = itemsDeUsuario[mascara].groupby('anio')['playtime_forever'].sum().sort_values(ascending=False)
    
    if not sumaDehoras.empty:
        anio = sumaDehoras.head(1).index[0]
    else:
        anio = None

    df_f1.loc[df_f1['genero'] == genero, 'anio'] = anio

    aniosYhoras = itemsDeUsuario[(itemsDeUsuario['anio'] == anio) & mascara].groupby('anio')['playtime_forever'].sum()
    total_anio = aniosYhoras.index
    horasJugadas = {}
    for horas, total_anio in enumerate(total_anio):
        horasJugadas[f'AÃ±o {total_anio}'] = f'Horas: {aniosYhoras.iloc[horas]}'

    df_f1.loc[df_f1['genero'] == genero, 'Horas jugadas'] = [horasJugadas]
```
