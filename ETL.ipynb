** ETL Steam Games **

```python


filas_games = []
with open("json/output_steam_games.json") as f:
    for line in f.readlines():
        data = json.loads(line)
        filas_games.append(data)


df_games = pd.DataFrame(filas_games)
df_games
```

```python
df_games.isnull().sum()
```

```python
# Eliminar registros con valores NaN en todos los campos
df_games.dropna(how='all', inplace=True)
```

```python
df_games.shape
```

```python
# Borrar columnas innecesarias
df_games.drop(['app_name','early_access','url','reviews_url','tags','publisher','specs'], axis=1, inplace = True)
df_games.reindex()
```

```python
df_games.shape
```

```python
df_games.isna().sum()
```

```python
#Eliminamos valores duplicados y nulos
df_games.drop_duplicates(subset='id', inplace=True) 
df_games.dropna(subset='id', inplace=True)
```

```python
df_games['anio'] = df_games['release_date'].str.extract(r'(\d{4})')
```

```python
df_games
```

```python
df_games.drop(columns=['release_date'], inplace = True)
df_games
```

```python
df_games= df_games.dropna(subset=['anio'])
df_games= df_games.dropna(subset=['genres'])
```

```python
df_games.head(2)
```

```python
df_games['price'].unique()
```

```python
# Reemplazar los valores que comienzan con "free" o "Free" por 0

def transformar_valor(valor):
    if pd.notna(valor):
        # Verificar si el valor es un nÃºmero
        try:
            valor_flotante = round(float(valor), 2)
            return valor_flotante
        except ValueError:
            # Si no es un nÃºmero, verificar si comienza con 'Free'
            if valor.lower().startswith('free'):
                return 0.0
    return 0

df_games['price'] = df_games['price'].apply(transformar_valor)
```

```python
df_games['price'].unique()
```

```python
df_games = df_games[pd.to_numeric(df_games['price'], errors='coerce').notnull()]
```

```python
df_games.to_parquet("data/df_games.parquet")
```

# ETL REVIEWS

```python
filas_reviews = []

with open("json/australian_user_reviews.json", encoding='MacRoman') as l:
    for linea in l.readlines():
        filas_reviews.append(ast.literal_eval(linea))

df_reviews = pd.DataFrame(filas_reviews)
```

```python
df_reviews
```

```python
df_reviews.isnull().sum()
```

```python
#Desanidar
df_re1 = df_reviews.explode(['reviews'])
df_re2 = df_re1['reviews'].apply(pd.Series)
df_re3 = pd.concat([df_re1, df_re2], axis=1)
```

```python
df_re3
```

```python

# Creamos una columna con el aÃ±o de la columna 'posted'
df_re3['year_posted'] = df_re3['posted'].str.extract('(\d{4})')

# Elimino columnas que no voy usar
df_re3.drop(['reviews','user_url','last_edited','funny','helpful','posted'], axis=1, inplace = True)

# Elimino valores nulos por columna
df_re3 = df_re3.dropna(subset=['year_posted'])
df_re3 = df_re3.dropna(subset=['item_id'])
df_re3 = df_re3.dropna(subset=['review'])
df_re3 = df_re3.dropna(subset=['recommend'])

# Me quedo con las columnas importantes
df_re3 = df_re3[['user_id','item_id','review','recommend','year_posted',0]]

# Para la recomendacion reemplazo valores
df_re3['recommend'] = df_re3['recommend'].replace({'False': 0, 'True': 1}).astype(int)

# Borro la ultima columna [0]
del df_re3[df_re3.columns[-1]]

# Funcion para analizar el sentimiento con la libreria TextBlob
def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment < -0.1:
        return 0
    elif sentiment > 0.1:
        return 2
    else:
        return 1

# Nueva columna 'sentiment_analysis'
df_re3['sentiment_analysis'] = df_re3['review'].apply(get_sentiment)

# Eliminar la columna Review
df_re3.drop(['review'], axis=1, inplace = True)


df_re3 = df_re3.drop_duplicates()
df_re3 = df_re3.dropna()


df_re3.to_csv('data/user_reviews.csv', index=False)
```
